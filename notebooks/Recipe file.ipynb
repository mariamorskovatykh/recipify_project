```python
import numpy as np
import pandas as pd
```

```python
recipe_df = pd.read_csv('/Users/malorycorteso/code/mariamorskovatykh/recipify/raw_data/RAW_recipes.csv')
recipe_df.head()
```

```python
#removing white spaces
recipe_df.columns = recipe_df.columns.str.replace(' ', '')
```

```python
#drop useless columns
recipe_df.drop(columns=['contributor_id', 'submitted', 'description'], inplace=True)
```

```python
#cleaning data
recipe_df['ingredients'] = recipe_df['ingredients'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
recipe_df['tags'] = recipe_df['tags'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
recipe_df['nutrition'] = recipe_df['nutrition'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
recipe_df['steps'] = recipe_df['steps'].map(lambda x: x.lstrip('[').rstrip(']').replace("'", ""))
```

```python
#getting nutrition data in several columns(calories, sugar...)
recipe_df['calories'] = recipe_df['nutrition'].str.split(',').str[0]
recipe_df['total fat'] = recipe_df['nutrition'].str.split(',').str[1]
recipe_df['sugar'] = recipe_df['nutrition'].str.split(',').str[2]
recipe_df['sodium'] = recipe_df['nutrition'].str.split(',').str[3]
recipe_df['protein'] = recipe_df['nutrition'].str.split(',').str[4]
recipe_df['saturated fat'] = recipe_df['nutrition'].str.split(',').str[5]
recipe_df['carbohydrates'] = recipe_df['nutrition'].str.split(',').str[6]
```

```python
recipe_df.drop(columns=['nutrition'], inplace=True)
```

```python
recipe_df = recipe_df.rename(columns={'id': 'recipe_id'})
```

```python
review_df = pd.read_csv('/Users/malorycorteso/code/mariamorskovatykh/recipify/raw_data/RAW_interactions.csv')
len(review_df)
```

```python
#checking duplicates
duplicate_count_review = len(review_df) - len(review_df.drop_duplicates()) 
#removing white spaces
review_df.columns = review_df.columns.str.replace(' ', '')
```

```python
review_df.head()
```

```python
#removing useless columns for reviews
review_df.drop(columns=['user_id', 'date', 'review'], inplace=True)
```

```python
review_df.head(5)
```

```python
avg_ratings = round(review_df.groupby('recipe_id')['rating'].mean(),2)
```

```python
#adding the average rating in the recipe dataframe
recipe_df = pd.merge(recipe_df, avg_ratings, on = ['recipe_id'])
recipe_df.head()
```

```python
# list of recipes from the image dataset
list_recipe = pd.read_fwf('/Users/malorycorteso/code/mariamorskovatykh/recipify/raw_data/class_list.txt', header=None)
list_recipe.columns =['Name']
list_recipe['Name'] = list_recipe['Name'].str.replace('\d+', '')
list_recipe['Name'] = list_recipe['Name'].str.replace('_', ' ')
list_recipe.index.name = 'label'
list_recipe.head()
```

```python
list_recipe.shape
```

```python
#check how many recipes in the image dataset are in Food.com Recipes
list_recipes = []
for recipe in list_recipe['Name']:
    recipe = recipe.replace('food', '')
    recipe = recipe.replace('macaron', 'macarons')
    recipe = recipe.strip()
    if recipe_df['name'].str.contains(recipe).any():
        list_recipes.append(recipe)
list_recipes
```

```python
list_recipes2 = []
for recipe in list_recipe['Name']:
    recipe = recipe.strip('food').strip()
    if ~recipe_df['name'].str.contains(recipe).any():
        list_recipes2.append(recipe)
list_recipes2
```

```python
#list of recipe with our labels from images dataset
pattern = '|'.join(list_recipes)
new_recipe_df = recipe_df[recipe_df['name'].str.contains(pattern)==True]
new_recipe_df.shape
```
